```python
months = np.arange(1, 145)
production = np.array(data)

plt.figure(figsize=(14, 7))
plt.plot(months, production, label='Monthly Production')
plt.xlabel('Month')
plt.ylabel('Production')
plt.title('Monthly Production of Bags (January 2018 - December 2023)')
plt.legend()
plt.grid(True)
plt.show()

coefficient = np.polyfit(months, production, 3)
poly = np.poly1d(coefficient)
prediction = poly(months)

plt.figure(figsize=(14, 7))
plt.plot(months, production, label='Actual Production')
plt.plot(months, prediction, label='Polynomial Trend', color='red')
plt.xlabel('Month')
plt.ylabel('Production')
plt.title('Monthly Production of Bags with Polynomial Trend')
plt.legend()
plt.grid(True)
plt.show()

a3, a2, a1, a0 = coefficient

print(f"Polynomial Coefficient: x = {a3:.3f}; x = {a2:.3f}; x = {a1:.3f}; constant = {a0:.3f}")
print(f"Mathematical Model    : P(x) = {a3:.3f}x + {a2:.3f}x + {a1:.3f}x + {a0:.3f}")
```

```python
a3, a2, a1, a0 = coefficient

def polynomialModel(x):
    return a3*x**3 + a2*x**2 + a1*x + a0

exMonth = 100
print(f"Prediction for month {exMonth}: {polynomialModel(exMonth):.3f}")
```

**Problem - 3**

```python
def equation(x):
    return polynomialModel(x) - 25000

root = fsolve(equation, 100)[0]  
buildMonth = root - 13

def convertDate(month):
    year = int(2018 + (month - 1) / 12)
    monthNumber = int((month - 1) % 12 + 1)
    monthName = calendar.month_name[monthNumber]
    return f"{monthName} {year}"

rootDate = convertDate(int(root))
buildDate = convertDate(int(buildMonth))

print(f"Month to start building new warehouse   : {buildDate}")
print(f"Month when production exceeds 25000 bags: {rootDate}")
```
